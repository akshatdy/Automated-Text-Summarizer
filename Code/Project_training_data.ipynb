```python
data_abs = os.path.join(data_path, 'IN-Abs')
```

```python
train_abs = os.path.join(data_abs, 'train-data')
```

```python
train_abs
```

```python
train_abs_judgement = os.path.join(train_abs, 'judgement')
train_abs_summary = os.path.join(train_abs, 'summary')
```

```python
train_abs_summary
```

```python
judgment_ids = []
judgments = []
summaries = []
```

```python
# Iterate over the files in the 'judgement' folder
for file in os.listdir(train_abs_judgement):
    if file.endswith('.txt'):
        
        # Extract the judgment ID from the file name
        judgment_id = int(file.split('.')[0])
        
        judgment_path = os.path.join(train_abs_judgement, file)
        with open(judgment_path, 'r') as f:
            judgment = f.read()
            judgment = judgment.replace('\n', ' ')
            judgment = judgment.replace("\'", '')
            judgment = judgment.replace('"', '')

        # Get the corresponding summary file name
        summary_file = f"{judgment_id}.txt"

        summary_path = os.path.join(train_abs_summary, summary_file)
        with open(summary_path, 'r') as f:
            summary = f.read()
            summary = summary.replace('\n', ' ')
            summary = summary.replace("\'", '')
            summary = summary.replace('"', '')

            summaries.append(summary)

        judgments.append(judgment)

        judgment_ids.append(judgment_id)
```

```python
# Create the dataframe
data = {'Judgment_ID': judgment_ids, 'Judgment': judgments, 'Summary': summaries}
df = pd.DataFrame(data)
```

```python
# Print the dataframe
df
```

```python
df = df.sort_values(by='Judgment_ID', ascending=True)
df
```

```python
# Remove rows with '~' in 'Judgment' or 'Summary' columns
df = df[~(df['Judgment'].str.contains('\t') | df['Summary'].str.contains('\t'))].copy()
df
```

```python
print(df.iloc[1][2])
```

```python
# Save the dataframe as a CSV file
df.to_csv(os.path.join(data_path, 'training_data_abs.csv'), index=False, sep='\t')
```

```python
# Save the dataframe as a XLSX file
df.to_excel(os.path.join(data_path, 'training_data_abs.xlsx'), index=False)
```
